```python
data=pd.read_csv("homeprices_dummy.csv")
```

```python
dummy=pd.get_dummies(data["town"])
```

```python
dummy
```

```python
data1=pd.concat([data,dummy],axis=1)
```

```python
data1
```

```python
data1.drop(["town"],axis=1,inplace=True)
data1
```

```python
x=data1.drop(["price"],axis=1)
y=data1["price"]
             
```
